```python
import os, sys
sys.path.append('..')

import warnings
warnings.filterwarnings('ignore', category=UserWarning, module='cobra')

from cobra import io
from scripts.helpers.model import rxn_in_model, add_single_gene_reaction_pair
from scripts.opt._fva import run_flux_variability_analysis

model = io.read_sbml_model('../data/fill/xmls/MNL_iCre1355_auto_GAPFILL.xml')

ERGEXCH = 'ERGOSTEROLEXCH'
ergosterol_c = 'ergosterol_c'
add_single_gene_reaction_pair(
    model=model, 
    gene_id="EXCHERG_GENE",
    reaction_id=ERGEXCH,
    reaction_name="Ergosterol exchange (assumption)", 
    reaction_subsystem="Exchange", 
    metabolites=[(-1, ergosterol_c)],
    reversible=True
)

reactions = [
    ERGEXCH, 'SS', 'SMO', 'HDS', 'IDS1', 'IDS2', 'IDIh', 'FPPSh', 'GPPSh', 'DXS', 'CMS', 'CMK', 'MCS', 'DXRI', 'ERG',
    'FRDPth', 'IPDPth', 'GRDPth'
]

fva = run_flux_variability_analysis(
    model,
    loopless=True,
    pfba_factor=None,
    objectives=[ERGEXCH],
    reactions=reactions
)

fva
```
